```python
#Visualization for BLS total employed data
groups = bls_data.groupby('Occupation')

for name, group in groups:
    plt.plot(group['Year'], group['Total Employed'], label=name)

plt.xlabel('Year')
plt.ylabel('Total Employed (Hundred Thousand)')
plt.legend()
plt.title('Historical Occupation Data')

plt.show()
```

The supply of Computer Network Architects has been gradually decreasing from the year 2012 to 2022, with the highest number of employed personnel in 2013. Whereas, the number of professionals working as Information Security Analyst has been increasing over the years, with a sudden spike around 2019 and the highest in the year 2022.

```python
#Implement the Augmented Dickey-Fuller (ADF) test to test data stationarity
```

```python
from statsmodels.tsa.stattools import adfuller
passing_data=adfuller(bls_data['Total Employed'])
```

```python
def adf_test(employed):
    result=adfuller(employed)
    labels = ['Test parameters', 'p-value','#Lags Used','Dataset observations']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    if result[1] <= 0.05:
        print("Dataset is stationary")
    else:
        print("Dataset is non-stationary ")
```

```python
adf_test(bls_data['Total Employed'])
```

```python
#Implement differencing to make the data stationary
```

```python
bls_data['Differencing']=bls_data['Total Employed']-bls_data['Total Employed'].shift(12)
```

```python
adf_test(bls_data['Differencing'].dropna())
```
